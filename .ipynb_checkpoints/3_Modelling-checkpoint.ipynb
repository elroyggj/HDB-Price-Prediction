```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime

pd.options.display.max_columns = None

%matplotlib inline

all_data = pd.read_csv('./datasets/merged_dataset_part1.csv', index_col=0).append(pd.read_csv('./datasets/merged_dataset_part2.csv', index_col=0))

# Keep only required features
all_data = all_data[['Month Sold', 'Town', 'Flat Type', 'Flat Model', 'Storey Range', 'Floor Area Sqm', 'Remaining Lease', 'Distance To Nearest Station', 'No. Of Stations Within 500M', 'Primary Distance', 'Primary Distance Categorised', 'Secondary Distance Categorised', 'Resale Price', 'Deflated Resale Price']]
all_data['Month Sold'] = all_data['Month Sold'].astype(str)
all_data['Additional Primary Distance Categorised'] = all_data['Primary Distance'].apply(lambda x: '<= 500m' if x <= 500 else ('501m - 1km' if x <= 1000 else ('1km - 1.5km' if x <= 1500 else ('1.5km - 2km' if x <= 2000 else '> 2km'))))
all_data.drop(columns=['Primary Distance'], inplace=True)
```

```python
from sklearn.ensemble import AdaBoostRegressor, RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.linear_model import ElasticNet, ElasticNetCV, Lasso, LassoCV, LinearRegression, Ridge, RidgeCV
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# Preparing X and y variables and separating them into train and test sets.

X = all_data[[col for col in all_data.columns if col not in ['Resale Price', 'Deflated Resale Price']]].copy()
y = all_data[['Resale Price']]
y_deflated = all_data[['Deflated Resale Price']]

ss = StandardScaler()

# X variables for Linear Regression (with first dummy variable dropped)
X_lr_dum = pd.get_dummies(X, drop_first=True)
Xs_lr_dum = pd.DataFrame(ss.fit_transform(X_lr_dum), index=X_lr_dum.index, columns=X_lr_dum.columns)


# X variables for other models (w all dummy variable)
X_dum = pd.get_dummies(X)
Xs_dum = pd.DataFrame(ss.fit_transform(X_dum), index=X_dum.index, columns=X_dum.columns)

reference_dum = [col for col in pd.get_dummies(X).columns if col not in X_lr_dum.columns]

X_lr_train, X_lr_test, y_train, y_test = train_test_split(Xs_lr_dum, y, test_size=0.1, random_state=42)

X_train = Xs_dum.loc[X_lr_train.index, :]
X_test = Xs_dum.loc[X_lr_test.index, :]

y_deflated_train = y_deflated.loc[y_train.index, :]
y_deflated_test =  y_deflated.loc[y_test.index, :]
```

```python
reference_dum
```

## Baseline Model (Linear Regression)

#### Linear Regression

```python
lr = LinearRegression()
score = cross_val_score(lr, X_lr_train, y_train, cv=5, n_jobs=-1)

lr.fit(X_lr_train, y_train)
lr_beta_coef = pd.DataFrame(lr.coef_, columns=X_lr_train.columns, index=['Linear Regression Beta']).T


print('='*35)
print('Linear Regression (Resale Price)')
print('='*35)
print('')
print('CV Scores:\t{}'.format(score))
print('Mean CV Scores:\t{}'.format(np.mean(score)))
print('')
print('Training R2:\t{}'.format(lr.score(X_lr_train, y_train)))
print('Test R2:\t{}'.format(lr.score(X_lr_test, y_test)))
print('')

residual = y_test - lr.predict(X_lr_test)

fig, ax = plt.subplots(figsize=(15, 8))
plt.scatter(x=lr.predict(X_lr_test), y=residual.values, alpha=0.1)
plt.title('\nResidual Plot of Linear Regression (Resale Price)\n')
plt.ylabel('Residual (Resale Price)\n')
plt.xlabel('\nFitted Values')
ax.axhline(c='r', lw=1, ls='dashed')
plt.show()

print('Residual Plots seems to reflect that the model is underfitted. Shall try deflated resale price as target variable.')
```

```python
lr = LinearRegression()
score = cross_val_score(lr, X_lr_train, y_deflated_train, cv=5, n_jobs=-1)

lr.fit(X_lr_train, y_deflated_train)
lr_beta_coef =pd.DataFrame(lr.coef_, columns=X_lr_train.columns, index=['Linear Regression Beta']).T

title = 'Linear Regression (Deflated Resale Price)'
print('='*len(title))
print(title)
print('='*len(title))
print('')
print('CV Scores:\t{}'.format(score))
print('Mean CV Scores:\t{}'.format(np.mean(score)))
print('')
print('Training R2:\t{}'.format(lr.score(X_lr_train, y_deflated_train)))
print('Test R2:\t{}'.format(lr.score(X_lr_test, y_deflated_test)))
print('')


residual = y_deflated_test - lr.predict(X_lr_test)

fig, ax = plt.subplots(figsize=(15, 8))
plt.scatter(x=lr.predict(X_lr_test), y=residual.values, alpha=0.1)
plt.title('\nResidual Plot of Linear Regression (Deflated Resale Price)\n')
plt.ylabel('Residual (Deflated Resale Price)\n')
plt.xlabel('\nFitted Values')
ax.axhline(c='r', lw=1, ls='dashed')
plt.show()

print('Residual Plots still shows that the model has heteroscedasticity - as fitted values increase, the residuals seems \
to increase as well. This is likely due to the fact that there are lower number of high value resale cases.')
```

#### Ridge Regression

```python
ridge_alphas = np.logspace(-2, 7, 50)
optimal_ridge = RidgeCV(alphas=ridge_alphas, cv=5)
optimal_ridge.fit(X_lr_train, y_deflated_train)

ridge = Ridge(alpha=optimal_ridge.alpha_)
score = cross_val_score(ridge, X_lr_train, y_deflated_train, cv=5, n_jobs=-1)

ridge = Ridge(alpha=optimal_ridge.alpha_)
ridge.fit(X_lr_train, y_deflated_train)
ridge_beta_coef = pd.DataFrame(ridge.coef_, columns=X_lr_train.columns, index=['Ridge Beta']).T


title = 'Ridge Regression (Deflated Resale Price)'
print('='*len(title))
print(title)
print('='*len(title))
print('')
print('Optimal Alpha:\t{}'.format(optimal_ridge.alpha_))
print('')
print('CV Scores:\t{}'.format(score))
print('Mean CV Scores:\t{}'.format(np.mean(score)))
print('')
print('Training R2:\t{}'.format(ridge.score(X_lr_train, y_deflated_train)))
print('Test R2:\t{}'.format(ridge.score(X_lr_test, y_deflated_test)))
print('')


residual = y_deflated_test - ridge.predict(X_lr_test)

fig, ax = plt.subplots(figsize=(15, 8))
plt.scatter(x=ridge.predict(X_lr_test), y=residual.values, alpha=0.1)
plt.title('\nResidual Plot of Ridge Regression (Deflated Resale Price)\n')
plt.ylabel('Residual (Deflated Resale Price)\n')
plt.xlabel('\nFitted Values')
ax.axhline(c='r', lw=1, ls='dashed')
plt.show()

print('Train and test sets R2 scores are not much better than vanilla linear regression despite including ridge \
regularisation.')
```

#### Lasso Regression

```python
optimal_lasso = LassoCV(n_alphas=500, cv=5, n_jobs=-1, tol=0.002, max_iter=10000)
optimal_lasso.fit(X_lr_train, y_deflated_train)

lasso = Lasso(alpha=optimal_lasso.alpha_)
score = cross_val_score(lasso, X_lr_train, y_deflated_train, cv=5, n_jobs=-1)

lasso = Lasso(alpha=optimal_lasso.alpha_)
lasso.fit(X_lr_train, y_deflated_train)
lasso_beta_coef = pd.DataFrame(lasso.coef_, index=X_lr_train.columns, columns=['Lasso Beta'])


title = 'Lasso Regression (Deflated Resale Price)'
print('='*len(title))
print(title)
print('='*len(title))
print('')
print('Optimal Alpha:\t{}'.format(optimal_lasso.alpha_))
print('')
print('CV Scores:\t{}'.format(score))
print('Mean CV Scores:\t{}'.format(np.mean(score)))
print('')
print('Training R2:\t{}'.format(lasso.score(X_lr_train, y_deflated_train)))
print('Test R2:\t{}'.format(lasso.score(X_lr_test, y_deflated_test)))
print('')


residual = y_deflated_test.values - pd.DataFrame(lasso.predict(X_lr_test)).values

fig, ax = plt.subplots(figsize=(15, 8))
plt.scatter(x=pd.DataFrame(lasso.predict(X_lr_test)).values, y=residual, alpha=0.1)
plt.title('\nResidual Plot of Lasso Regression (Deflated Resale Price)\n')
plt.ylabel('Residual (Deflated Resale Price)\n')
plt.xlabel('\nFitted Values')
ax.axhline(c='r', lw=1, ls='dashed')
plt.show()

print('Similarly, train and test sets R2 scores are not much better than vanilla linear regression despite including lasso \
regularisation.')
```

## Random Forest Regressor

```python
rf = RandomForestRegressor(n_estimators=10, n_jobs=-1)
score = cross_val_score(rf, X_train, y_deflated_train, cv=5, n_jobs=-1)

rf.fit(X_train, y_deflated_train)


title = 'Random Forest (Deflated Resale Price)'
print('='*len(title))
print(title)
print('='*len(title))
print('')
print('CV Scores:\t{}'.format(score))
print('Mean CV Scores:\t{}'.format(np.mean(score)))
print('')
print('Training R2:\t{}'.format(rf.score(X_train, y_deflated_train)))
print('Test R2:\t{}'.format(rf.score(X_test, y_deflated_test)))
print('')


residual = y_deflated_test.values - pd.DataFrame(rf.predict(X_test)).values

fig, ax = plt.subplots(figsize=(15, 8))
plt.scatter(x=pd.DataFrame(rf.predict(X_test)).values, y=residual, alpha=0.1)
plt.title('\nResidual Plot of Random Forest (Deflated Resale Price)\n')
plt.ylabel('Residual (Deflated Resale Price)\n')
plt.xlabel('\nFitted Values')
ax.axhline(c='r', lw=1, ls='dashed')
plt.show()

print('Residual plot of Random Forest seems to be more "even" than the linear regression models - Even the areas where \
predicted values are lower has several cases where residuals are high. Though overall, residuals does concentrate along a \
rather tight and even band.')
```

```python
rf_feature_importance = pd.DataFrame(rf.feature_importances_, columns=['Importance'], index=X_train.columns)
rf_feature_importance.to_csv('./Random_Forest_Feature_Importance.csv')
rf_feature_importance.sort_values(by='Importance', ascending=False)
```
