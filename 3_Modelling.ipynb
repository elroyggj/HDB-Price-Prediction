```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime

pd.options.display.max_columns = None

%matplotlib inline

all_data = pd.read_csv('./datasets/merged_dataset_part1.csv', index_col=0).append(pd.read_csv('./datasets/merged_dataset_part2.csv', index_col=0))

# Keep only required features
all_data = all_data[['Month Sold', 'Town', 'Flat Type', 'Flat Model', 'Storey Range', 'Floor Area Sqm', 'Remaining Lease', 'Distance To Nearest Station', 'No. Of Stations Within 500M', 'Primary Distance', 'Primary Distance Categorised', 'Secondary Distance Categorised', 'Resale Price', 'Deflated Resale Price']]
all_data['Month Sold'] = all_data['Month Sold'].astype(str)
all_data['Additional Primary Distance Categorised'] = all_data['Primary Distance'].apply(lambda x: '<= 500m' if x <= 500 else ('501m - 1km' if x <= 1000 else ('1km - 1.5km' if x <= 1500 else ('1.5km - 2km' if x <= 2000 else '> 2km'))))
all_data.drop(columns=['Primary Distance'], inplace=True)
```

```python
from sklearn.ensemble import AdaBoostRegressor, RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.linear_model import ElasticNet, ElasticNetCV, Lasso, LassoCV, LinearRegression, Ridge, RidgeCV
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import MinMaxScaler
```

```python
# Preparing X and y variables and separating them into train and test sets.

X = all_data[[col for col in all_data.columns if col not in ['Resale Price', 'Deflated Resale Price']]].copy()
y = all_data[['Resale Price']]
y_deflated = all_data[['Deflated Resale Price']]

mms = MinMaxScaler()

# X variables for Linear Regression (with first dummy variable dropped)
X_lr_dum = pd.get_dummies(X, drop_first=True)
Xs_lr_dum = pd.DataFrame(mms.fit_transform(X_lr_dum), index=X_lr_dum.index, columns=X_lr_dum.columns)


# X variables for other models (w all dummy variable)
X_dum = pd.get_dummies(X)
Xs_dum = pd.DataFrame(mms.fit_transform(X_dum), index=X_dum.index, columns=X_dum.columns)

reference_dum = [col for col in pd.get_dummies(X).columns if col not in X_lr_dum.columns]

X_lr_train, X_lr_test, y_train, y_test = train_test_split(Xs_lr_dum, y, test_size=0.1, random_state=42)

X_train = Xs_dum.loc[X_lr_train.index, :]
X_test = Xs_dum.loc[X_lr_test.index, :]

y_deflated_train = y_deflated.loc[y_train.index, :]
y_deflated_test =  y_deflated.loc[y_test.index, :]
```

## Baseline Model (Linear Regression)

#### Linear Regression

```python
lr = LinearRegression()
score = cross_val_score(lr, X_lr_train, y_train, cv=5, n_jobs=-1)

lr.fit(X_lr_train, y_train)
lr_beta_coef = pd.DataFrame(lr.coef_, columns=X_lr_train.columns, index=['Linear Regression Beta']).T


print('='*35)
print('Linear Regression (Resale Price)')
print('='*35)
print('')
print('CV Scores:\t{}'.format(score))
print('Mean CV Scores:\t{}'.format(np.mean(score)))
print('')
print('Training R2:\t{}'.format(lr.score(X_lr_train, y_train)))
print('Test R2:\t{}'.format(lr.score(X_lr_test, y_test)))
print('')

residual = y_test - lr.predict(X_lr_test)

fig, ax = plt.subplots(figsize=(15, 8))
plt.scatter(x=lr.predict(X_lr_test), y=residual.values, alpha=0.1)
plt.title('\nResidual Plot of Linear Regression (Resale Price)\n')
plt.ylabel('Residual (Resale Price)\n')
plt.xlabel('\nFitted Values')
ax.axhline(c='r', lw=1, ls='dashed')
plt.show()

print('Residual Plots seems to reflect that the model is underfitted. Shall try deflated resale price as target variable.')
```

```python
lr = LinearRegression()
score = cross_val_score(lr, X_lr_train, y_deflated_train, cv=5, n_jobs=-1)

lr.fit(X_lr_train, y_deflated_train)
lr_beta_coef =pd.DataFrame(lr.coef_, columns=X_lr_train.columns, index=['Linear Regression Beta']).T

title = 'Linear Regression (Resale Price)'
print('='*len(title))
print(title)
print('='*len(title))
print('')
print('CV Scores:\t{}'.format(score))
print('Mean CV Scores:\t{}'.format(np.mean(score)))
print('')
print('Training R2:\t{}'.format(lr.score(X_lr_train, y_deflated_train)))
print('Test R2:\t{}'.format(lr.score(X_lr_test, y_deflated_test)))
print('')


residual = y_deflated_test - lr.predict(X_lr_test)

fig, ax = plt.subplots(figsize=(15, 8))
plt.scatter(x=lr.predict(X_lr_test), y=residual.values, alpha=0.1)
plt.title('\nResidual Plot of Linear Regression (Deflated Resale Price)\n')
plt.ylabel('Residual (Deflated Resale Price)\n')
plt.xlabel('\nFitted Values')
ax.axhline(c='r', lw=1, ls='dashed')
plt.show()

print('Residual Plots still shows that the model may be underfitted and leads to heteroscedasticity - there seems to be \
a trend that as fitted values increase, residuals tends to become more positive. Shall try introducing exponents of some \
of the variables.')
```

```python
X['Area Square'] = X['Floor Area Sqm']**2
X['Remaining Lease Square'] = X['Remaining Lease']**2
X['No. Of Stations Within 500M Square'] = X['No. Of Stations Within 500M']**2

X_lr_dum = pd.get_dummies(X, drop_first=True)
Xs_lr_dum = pd.DataFrame(mms.fit_transform(X_lr_dum), index=X_lr_dum.index, columns=X_lr_dum.columns)

X_dum = pd.get_dummies(X)
Xs_dum = pd.DataFrame(mms.fit_transform(X_dum), index=X_dum.index, columns=X_dum.columns)

reference_dum = [col for col in pd.get_dummies(X).columns if col not in X_lr_dum.columns]

# X variables for Linear Regression (with first dummy variable dropped)
X_lr_train = Xs_lr_dum.loc[X_lr_train.index, :]
X_lr_test = Xs_lr_dum.loc[X_lr_test.index, :]


# X variables for other models (w all dummy variable)
X_train = Xs_dum.loc[X_lr_train.index, :]
X_test = Xs_dum.loc[X_lr_test.index, :]
```

```python
lr = LinearRegression()
score = cross_val_score(lr, X_lr_train, y_deflated_train, cv=5, n_jobs=-1)

lr.fit(X_lr_train, y_deflated_train)
lr_beta_coef =pd.DataFrame(lr.coef_, columns=X_lr_train.columns, index=['Linear Regression Beta']).T

title = 'Linear Regression (Deflated Resale Price)'
print('='*len(title))
print(title)
print('='*len(title))
print('')
print('CV Scores:\t{}'.format(score))
print('Mean CV Scores:\t{}'.format(np.mean(score)))
print('')
print('Training R2:\t{}'.format(lr.score(X_lr_train, y_deflated_train)))
print('Test R2:\t{}'.format(lr.score(X_lr_test, y_deflated_test)))
print('')


residual = y_deflated_test - lr.predict(X_lr_test)

fig, ax = plt.subplots(figsize=(15, 8))
plt.scatter(x=lr.predict(X_lr_test), y=residual.values, alpha=0.1)
plt.title('\nResidual Plot of Linear Regression (Deflated Resale Price)\n')
plt.ylabel('Residual (Deflated Resale Price)\n')
plt.xlabel('\nFitted Values')
ax.axhline(c='r', lw=1, ls='dashed')
plt.show()

print('Residual Plots shows that the trend still exists and there is still a mild U-shaped curve, and thus we shall \
attempt to introduce interaction terms between the variables. Cross validation score seems to be rather stable and does \
reflect that the model is unlikely to be overfitted. R2 score for both training and test sets have improved. We shall try \
to introduce interaction terms between Floor Area, Remaining Lease and No. of Station within 500M')
```

```python
X['Area X Remaining Lease'] = X['Floor Area Sqm'] * X['Remaining Lease']
X['Area X No. of Stations Within 500M'] = X['Floor Area Sqm'] * X['No. Of Stations Within 500M']
X['Remaining Lease X No. of Stations Within 500M'] = X['Remaining Lease'] * X['No. Of Stations Within 500M']

X_lr_dum = pd.get_dummies(X, drop_first=True)
Xs_lr_dum = pd.DataFrame(mms.fit_transform(X_lr_dum), index=X_lr_dum.index, columns=X_lr_dum.columns)

X_dum = pd.get_dummies(X)
Xs_dum = pd.DataFrame(mms.fit_transform(X_dum), index=X_dum.index, columns=X_dum.columns)

reference_dum = [col for col in pd.get_dummies(X).columns if col not in X_lr_dum.columns]

# X variables for Linear Regression (with first dummy variable dropped)
X_lr_train = Xs_lr_dum.loc[X_lr_train.index, :]
X_lr_test = Xs_lr_dum.loc[X_lr_test.index, :]


# X variables for other models (w all dummy variable)
X_train = Xs_dum.loc[X_lr_train.index, :]
X_test = Xs_dum.loc[X_lr_test.index, :]
```

```python
lr = LinearRegression()
score = cross_val_score(lr, X_lr_train, y_deflated_train, cv=5, n_jobs=-1)

lr.fit(X_lr_train, y_deflated_train)
lr_beta_coef =pd.DataFrame(lr.coef_, columns=X_lr_train.columns, index=['Linear Regression Beta']).T

title = 'Linear Regression (Deflated Resale Price)'
print('='*len(title))
print(title)
print('='*len(title))
print('')
print('CV Scores:\t{}'.format(score))
print('Mean CV Scores:\t{}'.format(np.mean(score)))
print('')
print('Training R2:\t{}'.format(lr.score(X_lr_train, y_deflated_train)))
print('Test R2:\t{}'.format(lr.score(X_lr_test, y_deflated_test)))
print('')


residual = y_deflated_test - lr.predict(X_lr_test)

fig, ax = plt.subplots(figsize=(15, 8))
plt.scatter(x=lr.predict(X_lr_test), y=residual.values, alpha=0.1)
plt.title('\nResidual Plot of Linear Regression (Deflated Resale Price)\n')
plt.ylabel('Residual (Deflated Resale Price)\n')
plt.xlabel('\nFitted Values')
ax.axhline(c='r', lw=1, ls='dashed')
plt.show()

print('Residual Plots shows that the trend still exists, though the U-Shape curve seems to be a lot less obvious. Cross \
validation score still seems to be rather stable which reflects that the model is unlikely to be overfitted. R2 score \
for both train and test sets have improved quite significantly from 0.67 to 0.87.')
```
